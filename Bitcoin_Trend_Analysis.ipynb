```
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import plotly.express as px
```

```
def GetStockData(tickerSymbol, start, end):
  #get data on this ticker
  tickerData = yf.Ticker(tickerSymbol)

  #get the historical prices for this ticker
  tickerDf = tickerData.history(period='1d', start=start, end=end)

  #see your data
  return tickerDf[['Close']]
```

```
def GetTrend(x,y):
  coef = np.polyfit(x,y,1)
  poly1d_fn = np.poly1d(coef) 
  return poly1d_fn(x), poly1d_fn
```

```
btc20 = GetStockData('BTC-USD', '2020-01-01', '2020-12-31')
btc21 = GetStockData('BTC-USD', '2021-01-01', '2021-12-31')

btc = pd.concat([btc20.tail(31), btc21])
```

```
trend, fn = GetTrend(np.arange(0,len(btc20)), btc20.values.reshape(-1))
trend1, fn1 = GetTrend(np.arange(len(btc20)-31, len(btc)+len(btc20)-31), btc.values.reshape(-1))

fig, axarr = plt.subplots(1,1,figsize=(16,8))

plt.title('Bitcoin')
plt.plot(btc20.index.values, btc20.values)
plt.plot(btc20.index.values, trend, 'r--')

plt.plot(btc21.index.values, btc21.values)
plt.plot(btc21.index.values, fn(np.arange(len(btc20), len(btc20)+len(btc21))), 'r--')

plt.plot(btc.index.values, trend1, 'g--')

plt.xticks(rotation=90)
plt.grid()
plt.savefig('Bitcoin.png')
```
