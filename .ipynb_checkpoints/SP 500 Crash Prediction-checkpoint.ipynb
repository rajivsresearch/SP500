```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv("SP5001.csv")
df2 = pd.read_csv("SP5002.csv")
df3 = pd.read_csv("SP5003.csv")
```

```python
print("DF1 : ", df1.shape)
print("DF2 : ", df2.shape)
print("DF3 : ", df3.shape)
```

```python
CPI = pd.read_csv("CPI.txt", sep="\t", header=0, encoding = "ISO-8859-1")
```

```python
CPI.drop(['Avg','Dec-Dec','Avg-Avg'], inplace=True, axis=1)
```

```python
CPI.iloc[-1, 6:] = 251.588
```

```python
CPI["July"] = CPI["July"].astype("float64")
CPI["Aug"] = CPI["Aug"].astype("float64")
CPI["Sep"] = CPI["Sep"].astype("float64")
CPI["Oct"] = CPI["Oct"].astype("float64")
CPI["Nov"] = CPI["Nov"].astype("float64")
CPI["Dec"] = CPI["Dec"].astype("float64")
```

```python
CPI.head()
```

```python
melted = pd.melt(CPI, id_vars=["Year"], 
                 var_name="Month", value_name="CPI")
```

```python
melted = melted.sort_values("Year")
```

```python
melted.reset_index(drop=True, inplace=True)
```

```python
melted.loc[:, "MonthNo"] = melted.Month.map({'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'June':6, 'July':7, 'Aug':8, 'Sep':9, 
                             'Oct':10, 'Nov':11, 'Dec':12})
```

```python
melted.loc[:,'MonthYear'] = melted.apply((lambda row: str(row['Year']) + '-' + str(row['MonthNo'])), axis=1)
```

```python
melted.set_index("MonthYear", inplace=True)
```

```python
melted.sort_index(inplace=True)
```

```python
melted.head()
```

```python
df = pd.concat([df1,df2,df3])
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.loc[:,"Date"] = df.loc[:,"Date"].astype("datetime64")

df.set_index("Date", inplace=True)

df.sort_index(inplace=True)
```

```python
df.describe()
```

```python
start = df.index.min()
end = df.index.max()
```

```python
plt.clf()
fig = plt.figure(figsize=(20,10))
ax = plt.subplot(1,1,1)
ax.set_title("Closing Price")

df.plot(x=df.index, y='Close', ax=ax)
plt.show()
```

```python
roi = [0.00]
for i in range(1,len(df)):
    curr_close = df.iloc[i,1]
    prev_close = df.iloc[i-1,1]
    ret = (prev_close - curr_close)/prev_close
    roi.append(ret)    
```

```python
df["DailyROI"] = roi
```

```python
df["DailyROI"].describe()
```

```python
df["year"] = df.index.year
```

```python
df["year_month"] = df.index.map(lambda x: str(x.year) + "-" + str(x.month))
```

```python
monthly_stats = pd.DataFrame(index=pd.unique(df.loc[:,"year_month"]))
```

```python
monthly_stats.head()
```

```python
dftemp = df.loc[:,["Close", "year_month"]]
```

```python
dftemp.head()
```

```python
dftemp.set_index('year_month', inplace=True)
```

```python
dftemp.head()
```

```python
groupvar = dftemp.loc[:,"Close"].groupby("year_month")
```

```python
monthly_stats["average"] = groupvar.mean()
```

```python
monthly_stats["stdev"] = groupvar.std()
```

```python
monthly_stats["max"] = groupvar.max()
monthly_stats["min"] = groupvar.min()
monthly_stats["median"] = groupvar.median()

monthly_stats["range"] = monthly_stats.loc[:,"max"] - monthly_stats.loc[:,"min"]
```

```python
monthly_stats.shape
```

```python
monthly_stats.head()
```

```python
monthly_stats["log"] = np.log(monthly_stats["average"])
```

```python
plt.clf()
fig = plt.figure(figsize=(20,10))
ax = plt.subplot(2,1,1)
ax.set_title("Monthly Stats - Average")

monthly_stats.plot(x=monthly_stats.index, y="average", ax=ax)

ax = plt.subplot(2,1,2)
ax.set_title("Monthly Stats - Log")

monthly_stats.plot(x=monthly_stats.index, y="log", ax=ax, c='r')

plt.show()
```

```python
from scipy.optimize import curve_fit
```

```python
def func(x, a, b, c):
    return a * np.exp(-b * x) + c
```

```python
xdata = np.arange(0, monthly_stats.shape[0])
```

```python
ydata = monthly_stats["log"].values
```

```python
plt.clf()
plt.plot(xdata, ydata, 'b-', label='data')
plt.show()
```

```python
#popt, pcov = curve_fit(func, xdata, ydata, maxfev=2000)
```

```python
def pfunc(x, a, b, c, d):
    return a * x ** 3 + b * x ** 2 + c * x + d
```

```python
z1 = np.polyfit(xdata, ydata, 1)
z2 = np.polyfit(xdata, ydata, 2)
z3 = np.polyfit(xdata, ydata, 3)
z4 = np.polyfit(xdata, ydata, 4)
z5 = np.polyfit(xdata, ydata, 5)
z6 = np.polyfit(xdata, ydata, 6)
```

```python
p1 = np.poly1d(z1)
p2 = np.poly1d(z2)
p3 = np.poly1d(z3)
p4 = np.poly1d(z4)
p5 = np.poly1d(z5)
p6 = np.poly1d(z6)

zdata1 = p1(xdata)
zdata2 = p2(xdata)
zdata3 = p3(xdata)
zdata4 = p4(xdata)
zdata5 = p5(xdata)
zdata6 = p6(xdata)
```

```python
plt.clf()

plt.figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')

plt.plot(xdata, ydata, 'b.')
plt.plot(xdata, zdata1, 'g.')
plt.plot(xdata, zdata2, 'r.')
plt.plot(xdata, zdata3, 'c.')
plt.show()
```

```python
plt.clf()

plt.figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')

plt.plot(xdata, ydata, 'b.')
plt.plot(xdata, zdata4, 'm.')
plt.plot(xdata, zdata5, 'y.')
plt.plot(xdata, zdata6, 'k.')
plt.show()
```

```python
avgbycpi = []

for item in monthly_stats.index.values:
    if melted.loc[item,"CPI"] > 0:
        avgprice = monthly_stats.loc[item, "average"]/melted.loc[item,"CPI"]
        avgbycpi.append(avgprice)
    else:
        #avgprice = monthly_stats.loc[item, "average"]/melted.loc[item,"CPI"]
        print(item, melted.loc[item,"CPI"])
```

```python
monthly_stats.loc[:,"avgbycpi"] = avgbycpi

#melted.loc[melted["CPI"] == 0, "CPI"] = 251.588
```

```python
monthly_stats.head()
```

```python
monthly_stats["logavgbycpi"] = np.log(monthly_stats["avgbycpi"])
```

```python
plt.clf()
fig = plt.figure(figsize=(20,20))
ax = plt.subplot(4,1,1)
ax.set_title("Monthly Stats - Average")

monthly_stats.plot(x=monthly_stats.index, y="average", ax=ax)

ax = plt.subplot(4,1,2)
ax.set_title("Monthly Stats - Log")

monthly_stats.plot(x=monthly_stats.index, y="log", ax=ax, c='r')

ax = plt.subplot(4,1,3)
ax.set_title("Monthly Stats - Average / CPI")

monthly_stats.plot(x=monthly_stats.index, y="avgbycpi", ax=ax, c='y')

ax = plt.subplot(4,1,4)
ax.set_title("Monthly Stats - Log Average / CPI")

monthly_stats.plot(x=monthly_stats.index, y="logavgbycpi", ax=ax, c='m')

plt.show()
```
