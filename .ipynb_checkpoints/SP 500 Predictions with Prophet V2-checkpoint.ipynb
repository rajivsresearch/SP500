```python
import pandas as pd
from fbprophet import Prophet

import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv("SP5001.csv")
df2 = pd.read_csv("SP5002.csv")
df3 = pd.read_csv("SP5003.csv")
```

```python
print("DF1 : ", df1.shape)
print("DF2 : ", df2.shape)
print("DF3 : ", df3.shape)
```

```python
df = pd.concat([df1,df2,df3])
```

```python
df.shape
```

```python
df.info()
```

```python
df.loc[:,"Date"] = df.loc[:,"Date"].astype("datetime64")

#df.set_index("Date", inplace=True)

#df.sort_index(inplace=True)
```

```python
df["Year"] = pd.DatetimeIndex(df['Date']).year
```

```python
df["Month"] = pd.DatetimeIndex(df['Date']).month
```

```python
df.head()
```

```python
CPI = pd.read_csv("CPI.txt", sep="\t", header=0, encoding = "ISO-8859-1")
```

```python
CPI.drop(['Avg','Dec-Dec','Avg-Avg'], inplace=True, axis=1)
```

```python
CPIM = pd.melt(CPI, id_vars=["Year"], 
                 var_name="Month", value_name="CPI")
```

```python
CPIM.loc[:, "MonthNo"] = CPIM.Month.map({'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'June':6, 'July':7, 'Aug':8, 'Sep':9, 
                             'Oct':10, 'Nov':11, 'Dec':12})
```

```python
CPIM.sort_values(["Year","MonthNo"], inplace=True)
```

```python
CPIM.reset_index(drop=True, inplace=True)
```

```python
CPIM.loc[CPIM.CPI == ' ', "CPI"] = 252
```

```python
CPIM.loc[:,"CPI"] = CPIM.loc[:,"CPI"].astype("float64")
```

```python
df2 = pd.merge(df, CPIM, how='left', left_on=['Year','Month'], right_on=['Year','MonthNo'])
```

```python
df2.sort_values("Date", inplace=True)
```

```python
df2.reset_index(drop=True, inplace=True)
```

```python
df2["Price"] = df2["Close"] / df2["CPI"]
```

```python
df2["Price100"] = df2["Price"].rolling(100).mean()
```

```python
df2["Price200"] = df2["Price"].rolling(200).mean()
```

```python
df2["LogPrice200"] = np.log(df2["Price200"])
```

```python
df2.iloc[95:,:].head(10)
```

```python
df2.plot("Date", ["LogPrice200"], figsize=(20,10))
```

```python
profdf = pd.DataFrame(data=df2.loc[df2["Year"]>1988 ,["Date","LogPrice200"]])
```

```python
profdf2 = pd.DataFrame(data=df2.loc[df2["Year"]>1988 ,["Date","CPI"]])
```

```python
profdf.reset_index(inplace=True, drop=True)
```

```python
profdf2.reset_index(inplace=True, drop=True)
```

```python
profdf.rename(columns={"Date":"ds","LogPrice200":"y"}, inplace=True)
```

```python
profdf2.rename(columns={"Date":"ds","CPI":"y"}, inplace=True)
```

```python
m = Prophet()
```

```python
m2 = Prophet()
```

```python
m.fit(profdf)
```

```python
m2.fit(profdf2)
```

```python
future = m.make_future_dataframe(periods=3650)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
forecast2 = m2.predict(future)
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m2.plot(forecast2)
```

```python
pred = forecast.iloc[profdf.shape[0]:,:]
```

```python
pred2 = forecast2.iloc[profdf2.shape[0]:,:]
```

```python
pred2.shape
```

```python
pred.plot("ds",["yhat", 'yhat_lower', 'yhat_upper'], figsize=(10,8))
```

```python
pred2.plot("ds",["yhat", 'yhat_lower', 'yhat_upper'], figsize=(10,8))
```

```python
from fbprophet.plot import add_changepoints_to_plot
fig = m.plot(forecast)
a = add_changepoints_to_plot(fig.gca(), m, forecast)
```

```python
pred["200dayma"] = np.exp(pred["yhat"])*pred2["yhat"]
```

```python
pred["200daymalower"] = np.exp(pred["yhat_lower"])*pred2["yhat"]
pred["200daymaupper"] = np.exp(pred["yhat_upper"])*pred2["yhat"]
```
