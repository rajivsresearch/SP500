```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import plotly.express as px
```

```
sp = pd.read_csv("/content/drive/MyDrive/Data/SP500/SP500.csv")
sp = sp.drop(columns=['Open', 'High', 'Low', 'Close', 'Volume'])
sp.columns = ['Date', 'Close']
sp['Date'] = pd.to_datetime(sp['Date'])
sp['DOY'] = sp['Date'].dt.dayofyear
sp['Year'] = sp['Date'].dt.year
sp.tail()
```

```
btc = pd.read_csv("/content/drive/MyDrive/Data/SP500/BTC-USD.csv")
btc = btc.drop(columns=['Open', 'High', 'Low', 'Close', 'Volume'])
btc.columns = ['Date', 'Close']
btc['Date'] = pd.to_datetime(btc['Date'])
btc['DOY'] = btc['Date'].dt.dayofyear
btc['Year'] = btc['Date'].dt.year
btc.tail()
```

```
df1 = sp.loc[sp['Year']== 2020, ['Date', 'Close']]
df1['Pct'] = (df1['Close'] - df1.iloc[0,1])/df1.iloc[0,1]*100
df1['Type'] = 'S&P Return'
df1['DOY'] = df1['Date'].dt.dayofyear
df1 = df1[['DOY', 'Type', 'Pct']].reset_index(drop=True)
df1.head()
```

```
df2 = btc.loc[btc['Year']== 2020, ['Date', 'Close']]
df2['Pct'] = (df2['Close'] - df2.iloc[0,1])/df2.iloc[0,1]*100
df2['Type'] = 'Bitcoin Return'
df2['DOY'] = df2['Date'].dt.dayofyear
df2 = df2[['DOY', 'Type', 'Pct']].reset_index(drop=True)
df2.head()
```

```
common = np.intersect1d(df1.DOY.values, df2.DOY.values) 
df3 = pd.concat([df1.loc[df1['DOY'].isin(common)], df2.loc[df2.DOY.isin(common)]])
df3.head()
```

```
fig = px.bar(df3, x='Type', y='Pct', animation_frame="DOY", range_y=[df3['Pct'].min()-10, df3['Pct'].max()+10])
fig.show()
```
