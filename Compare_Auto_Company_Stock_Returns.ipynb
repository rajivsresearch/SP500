```
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import plotly.express as px
```

```
def GetStockData(tickerSymbol, start, end):
  #get data on this ticker
  tickerData = yf.Ticker(tickerSymbol)

  #get the historical prices for this ticker
  tickerDf = tickerData.history(period='1d', start=start, end=end)

  #see your data
  return tickerDf[['Close']]
```

```
tsla = GetStockData('TSLA', '2020-01-01', '2020-12-31')
ford = GetStockData('F', '2020-01-01', '2020-12-31')
gm = GetStockData('GM', '2020-01-01', '2020-12-31')
fcau = GetStockData('FCAU', '2020-01-01', '2020-12-31')
```

```
df = pd.concat([tsla, ford, gm, fcau], axis=1)
df.columns = ['Tesla', 'Ford', 'GM', 'Fiat Chrysler']
```

```
df['DOY'] = df.index.dayofyear
```

```
df.head()
```

```
df1 = df.melt(id_vars = 'DOY', value_vars=['Tesla', 'Ford', 'GM', 'Fiat Chrysler'])
```

```
df1.columns = ['DOY', 'Name', 'Price']
```

```
f1 = df1['Name'] == 'Tesla'
df1.loc[f1,'Type'] = 'Tesla'
df1.loc[~f1,'Type'] = 'Ford, GM, Chrysler'
```

```
df1.head()
```

```
fig = px.bar(df1, x='Type', y='Price', animation_frame="DOY", animation_group = 'Name', color='Type')
fig.show()
```
