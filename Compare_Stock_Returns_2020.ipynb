```
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import plotly.express as px
```

```
def GetStockData(tickerSymbol, start, end):
  #get data on this ticker
  tickerData = yf.Ticker(tickerSymbol)

  #get the historical prices for this ticker
  tickerDf = tickerData.history(period='1d', start=start, end=end)

  #see your data
  return tickerDf[['Close']]
```

```
msft = GetStockData('MSFT', '2020-01-01', '2020-12-31')
goog = GetStockData('GOOG', '2020-01-01', '2020-12-31')
amzn = GetStockData('AMZN', '2020-01-01', '2020-12-31')
nflx = GetStockData('NFLX', '2020-01-01', '2020-12-31')
fb = GetStockData('FB', '2020-01-01', '2020-12-31')
aapl = GetStockData('AAPL', '2020-01-01', '2020-12-31')
#tsla = GetStockData('TSLA', '2020-01-01', '2020-12-31')
#zm = GetStockData('ZM', '2020-01-01', '2020-12-31')
```

```
fangam = pd.concat([fb, amzn, nflx, goog, aapl, msft], axis=1)
fangam.columns = ['fb', 'amzn', 'nflx', 'goog', 'aapl', 'msft']
```

```
i=0
df = pd.DataFrame(columns=fangam.columns)
for col in ['fb', 'amzn', 'nflx', 'goog', 'aapl', 'msft']:
  df[col] = (fangam[col] - fangam.iloc[0,i])/fangam.iloc[0,i]*100
  i=i+1
```

```
df['DOY'] = df.index.dayofyear
```

```
df.head()
```

```
df1 = df.melt(id_vars = 'DOY', value_vars=['fb', 'amzn', 'nflx', 'goog', 'aapl', 'msft'])
```

```
df1.columns = ['DOY', 'Ticker', 'Return']
```

```
df1.head()
```

```
fig = px.bar(df1, x='Ticker', y='Return', animation_frame="DOY", range_y=[df1['Return'].min()-10, df1['Return'].max()+10])
fig.show()
```
